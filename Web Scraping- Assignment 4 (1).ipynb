# Question 1:-

```python
driver = webdriver.Chrome()
```

```python
# Opening the Youtube page on Automated Chrome Browser.
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
data = []
Rank = []
Name = []
Artist = []
Upload_date = []
View = []
```

```python
tbody = driver.find_element(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[2]/tbody')
```

```python
for tr in tbody.find_elements(By.XPATH,'//tr'):
    row = [item.text for item in tr.find_elements(By.XPATH, './/td')]
    data.append(row)
print(data)
```

```python
# Scraping Rank from the given page
Sr = driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[2]/tbody/tr[1]')

for i in Sr:
    x = i.text
    Rank.append(x)
    
# Scraping Name from the given page.
name = driver.find_elements(By.XPATH,'href')

for i in name:
    naam = i.text
    Name.append(naam)
    
# Scraping Artist Name from the given page.
artist = driver.find_elements(By.XPATH,'href')

for i in artist:
    y = i.text
    Artist.append(y)
    
# Scraping upload date from the given page.
date = driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[2]/tbody/tr[1]/td[3]')

for i in date:
    dt = i.text
    Upload_date.append(dt)
    
# Scraping views from the given page.
view = driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[2]/tbody/tr[1]/td[4]')

for i in view:
    vw = i.text
    View.append(vw)
```

```python
driver = webdriver.Chrome()
driver.get("https://archive.ics.uci.edu/")
```

```python
datasets = driver.find_element(By.XPATH, '//div[@class="flex flex-wrap justify-center gap-5"]/a')
datasets.click()
```

```python
Dataset_name = []
Data_type = []
Task = []
Attribute_type = []
No_of_instances = []
No_of_attribute = []
Year = []
```

```python
# Scraping Dataset_name from the given page
dataset = driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')

for i in dataset:
    name = i.text
    Dataset_name.append(name)
    
# Scraping Data_type from the given page.
type = driver.find_elements(By.XPATH,'//div[@class="col-span-3 flex items-center gap-2"]/span')

for i in type:
    data = i.text
    Data_type.append(data)
    
# Scraping Task from the given page.
task = driver.find_elements(By.XPATH,'//span[@class="truncate"]')

for i in task:
    tk = i.text
    Task.append(tk)
    
# Scraping Attribute_type from the given page.
attribute = driver.find_elements(By.XPATH,'//table[@class="col-span-full my-2 table sm:col-start-2"]//td')

for i in attribute:
    at = i.text
    Attribute_type.append(at)
    
# Scraping No_of_instances from the given page.
instances = driver.find_elements(By.XPATH,'//span[@class="truncate"]')

for i in instances:
    noi = i.text
    No_of_instances.append(noi)
    
# Scraping No_of_attribute from the given page.
noa = driver.find_elements(By.XPATH,'//span[@class="truncate"]')

for i in noa:
    pr = i.text
    No_of_attribute.append(pr)
    
# Scraping Year from the given page.
year = driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div[1]/div[2]/div/table/tbody/tr/td[3]')

for i in year:
    yr = i.text
    Year.append(yr)
```
