```python
import selenium
import pandas as pd
from selenium import webdriver

import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
from time import sleep
```

```python
# Question 1:-

driver = webdriver.Chrome()
```

```python
# Opening the Shine page on Automated Chrome Browser.
driver.get("https://www.shine.com/")
```

```python
# Entering Designation as required in the question.

title = driver.find_element(By.XPATH, '//input[@class="form-control  "]')
title.send_keys("Data Analyst")
```

```python
# Entering Location as required in the question.
location = driver.find_element(By.XPATH, '/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input')
location.send_keys("Bangalore")
```

```python
search = driver.find_element(By.XPATH, '//button[@class=" btn btn-secondary undefined"]')
search.click()
```

```python
job_title = []
job_location = []
company_name = []
Experience_required = []
```

```python
# Scraping Job Title from the given page
title_tags = driver.find_elements(By.XPATH,'//h2[@itemprop="name"]/a')

for i in title_tags:
    title = i.text
    job_title.append(title)
    
# Scraping Job Location from the given page.
location_tags = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')

for i in location_tags:
    location = i.text
    job_location.append(location)
    
# Scraping Company Name from the given page.
company_tags = driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]/span')

for i in company_tags:
    company = i.text
    company_name.append(company)
    
# Scraping Experience required from the given page.
experience = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')

for i in experience:
    required = i.text
    Experience_required.append(required)
```

```python
print(len(job_title), len(job_location), len(company_name), len(Experience_required))
```

```python
df = pd.DataFrame({'Job Title':job_title, 'Job Location':job_location, 'Company Name':company_name, 'Experience_Required':Experience_required})
df
```

```python
# Opening the respected former Prime Ministers of India page on Automated Chrome Browser.
driver.get("http://www.jagranjosh.com/")
```

```python
search = driver.find_element(By.XPATH, '//a[@href="https://www.jagranjosh.com/general-knowledge"]')
search.click()
```

```python
search = driver.find_element(By.XPATH, '//a[@href="https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1"]')
search.click()
```

```python
Name = []
Born_Dead = []
Term_of_office = []
Remarks = []
```

```python
title = driver.find_element(By.XPATH, '//div[@class="TableData"]')

# Scraping Name from the given page
name = driver.find_elements(By.XPATH, '/html/body/div[1]/main/div[1]/div[1]/article/div[3]/div[5]/div/table/tbody/tr[2]/td[2]/div/strong/a')

for i in name:
    title = i.text
    Name.append(title)
    
# Scraping Born-Dead from the given page.
born = driver.find_elements(By.XPATH, '/html/body/div[1]/main/div[1]/div[1]/article/div[3]/div[5]/div/table/tbody/tr[2]/td[3]/div')

for i in born:
    dead = i.text
    Born_Dead.append(dead)
    
# Scraping Term of Office from the given page.
office = driver.find_elements(By.XPATH, '/html/body/div[1]/main/div[1]/div[1]/article/div[3]/div[5]/div/table/tbody/tr[2]/td[4]')

for i in office:
    term = i.text
    Term_of_office.append(term)
    
# Scraping  Remarks from the given page.
remarks = driver.find_elements(By.XPATH, '/html/body/div[1]/main/div[1]/div[1]/article/div[3]/div[5]/div/table/tbody/tr[2]/td[5]/div')

for i in remarks:
    mark = i.text
    Remarks.append(mark)
```

```python
print(len(Name), len(Born_Dead), len(Term_of_office), len(Remarks))
```

```python
df = pd.DataFrame({'Name':Name, 'Born-Dead':Born_Dead, 'Term of office':Term_of_office, 'Remarks':Remarks})
df
```
